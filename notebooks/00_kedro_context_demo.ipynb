# Kedro Context Demo

Cómo levantar el contexto de Kedro para acceder al catálogo de datos.

**Kernel:** `Python (credit-risk-frontier)`

Al cargar la extensión `kedro.ipython` quedan disponibles automáticamente:
- `catalog` — acceso a todos los datasets del catálogo
- `context` — contexto del proyecto
- `session` — sesión activa de Kedro
- `pipelines` — pipelines registrados

```python
%load_ext kedro.ipython
```

## Datasets disponibles en el catálogo

En Kedro 1.2.x se usa `catalog.keys()` (no `catalog.list()`).

```python
list(catalog.keys())
```

## Cargar dataset `pagos`

```python
pagos = catalog.load("pagos")
print(pagos.shape)
pagos.head()
```
